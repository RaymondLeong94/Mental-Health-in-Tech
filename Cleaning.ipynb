```python
df.work_interfere.value_counts().sum
```

```python
#this function creates a column for 0's and 1's to represent those afflicted
def mental_health_present(row):
    if row['work_interfere'] == 'No MI':
        return 0
    else:
        return 1

# Apply the function to create the new column
df['Mental_Health'] = df.apply(mental_health_present, axis=1)
```

```python
#Very fast function to organize gender/nonbinary
def categorize_gender(value):
    if value in ['woman', 'female', 'f']:
        return 'Female'
    elif value in ['m', 'male', 'make', 'man', 'cis male']:
        return 'Male'
    else:
        return 'Unknown'
df['0. Gender'] = df['0. Gender'].apply(categorize_gender)
```
